```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels as sm
import seaborn as sns
```

```python
df = pd.read_csv("all_data.csv")
print(df)
```

```python
for country in df.Country.unique():
    plt.plot(df.Year[df.Country == country], df["Life expectancy at birth (years)"][df.Country == country])
plt.legend(df.Country.unique())
plt.show()
plt.clf()
```

```python
for country in df.Country.unique():
    plt.plot(df.Year[df.Country == country], df["GDP"][df.Country == country])
plt.legend(df.Country.unique())
plt.show()
plt.clf()
```

```python
for country, i in zip(df.Country.unique(), range(1, 7)):
    a, b = np.polyfit(df.GDP[df.Country == country], df["Life expectancy at birth (years)"][df.Country == country], 1)
    plt.subplot(2, 3, i)
    plt.scatter(df.GDP[df.Country == country], df["Life expectancy at birth (years)"][df.Country == country])
    plt.plot(df.GDP[df.Country == country], (a) * (df.GDP[df.Country == country]) + b, alpha = 0.5, color="red")
    plt.title(f"Life Expectancy vs GDP in {country}")
    plt.xlabel("GDP")
    plt.ylabel("Life Expectancy")
plt.subplots_adjust(top=2.4, right=2.4)
plt.show()
plt.clf()
```

```python
sns.barplot(x="Country", y="GDP", data=df[df.Year == 2015], order=df[df.Year == 2015].sort_values(by="GDP")["Country"])
ax = plt.subplot()
ax.set_xticklabels(["Zimbabwe", "Chile", "Mexico", "Germany", "China", "USA"])
plt.title("2015 GDP Comparison")
plt.show()
plt.clf()
```
